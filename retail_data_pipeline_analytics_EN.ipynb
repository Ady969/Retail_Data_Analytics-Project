```python
import pandas as pd
import numpy as np

df = pd.read_csv('orders.csv', na_values=['Not Available', 'unknown'])
df.head(20)
```

```python
# 2. Removing spaces between column names, replacing them with underscore (_), and converting everything to lowercase
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.columns
```

```python
# Derive a new column: discount
df['discount']=df['list_price']*df['discount_percent']*.01
df.head()
```

```python
# New columns - sale_price and profit
df['sale_price']= df['list_price']-df['discount']
df['profit']=df['sale_price']-df['cost_price']
df
```

```python
# Converting the Order Date column into proper Date format
df['order_date'] = pd.to_datetime(df['order_date'], format="%Y-%m-%d")
df['order_date']
```

```python
# Drop cost price, list price and discount percent columns
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
df.columns
```

```python
# 2. 'retail_analytics.db' - local SQL database file
db_conn = duckdb.connect('retail_analytics.db')
```

```python
# 3. Register Pandas Dataframe in the form of SQL Table
db_conn.register('retail_sales', df)
```

```python
# retail_sales
db_conn.execute("""
    CREATE OR REPLACE TABLE fact_orders AS
    SELECT
        order_id,
        order_date,
        ship_mode,
        segment AS customer_segment,
        country,
        city,
        state,
        postal_code,
        region,
        category AS product_category,
        sub_category AS product_sub_category,
        product_id,
        quantity,
        discount AS discount_amount,
        sale_price AS sales,
        profit AS net_profit
    FROM retail_sales;
""")

print("⚙️ Data Infrastructure Set! 'fact_orders' table successfully created.")
```

**Query 1: Top 10 Revenue Generating Products**

```python
query_1 = """
SELECT product_id, ROUND(SUM(sales), 2) AS total_revenue
FROM fact_orders
GROUP BY product_id
ORDER BY total_revenue DESC
LIMIT 10;
"""

# Running the query and viewing the result in table form
df_res1 = db_conn.execute(query_1).df()
df_res1
```

**Query 2: Top 5 Highest Selling Products in Each Region**

```python
query_2 = """
WITH regional_sales AS (
    SELECT region, product_id, ROUND(SUM(sales), 2) AS total_sales,
           ROW_NUMBER() OVER(PARTITION BY region ORDER BY SUM(sales) DESC) AS rn
    FROM fact_orders
    GROUP BY region, product_id
)
SELECT region, product_id, total_sales
FROM regional_sales
WHERE rn <= 5;
"""

df_res2 = db_conn.execute(query_2).df()
df_res2
```

**Query-3 (Month-over-Month Growth Comparison for 2022 vs 2023)**

```python
query_3 = """
WITH monthly_metrics AS (
    SELECT
        EXTRACT(YEAR FROM order_date) AS order_year,
        EXTRACT(MONTH FROM order_date) AS order_month,
        SUM(sales) AS total_sales
    FROM fact_orders
    GROUP BY 1, 2
)
SELECT
    order_month,
    ROUND(SUM(CASE WHEN order_year = 2022 THEN total_sales ELSE 0 END), 2) AS sales_2022,
    ROUND(SUM(CASE WHEN order_year = 2023 THEN total_sales ELSE 0 END), 2) AS sales_2023,
    ROUND(((SUM(CASE WHEN order_year = 2023 THEN total_sales ELSE 0 END) - SUM(CASE WHEN order_year = 2022 THEN total_sales ELSE 0 END)) / NULLIF(SUM(CASE WHEN order_year = 2022 THEN total_sales ELSE 0 END), 0)) * 100, 2) AS yoy_growth_pct
FROM monthly_metrics
GROUP BY order_month
ORDER BY order_month;
"""

# Running the query and viewing the result in table form
df_res3 = db_conn.execute(query_3).df()
df_res3
```

**Query-4 (Highest Sales Month For Each Category)**

```python
query_4="""
WITH category_monthly_sales AS (
    SELECT
        product_category,
        STRFTIME(order_date, '%Y-%m') AS year_month, --Extract year and month from order_date
        ROUND(SUM(sales), 2) AS total_sales,
        ROW_NUMBER() OVER(PARTITION BY product_category ORDER BY SUM(sales) DESC) AS rn --step-2 rank dena
        FROM fact_orders
        GROUP BY product_category, STRFTIME(order_date, '%Y-%m')
    )
SELECT product_category, year_month, total_sales
FROM category_monthly_sales
WHERE rn = 1; -- Step 3: Sirf Rank 1 (Highest) filter karna
"""

# Running the query and viewing the result in table form
df_res4 = db_conn.execute(query_4).df()
df_res4
```

**Query-5(Sub-Category with Highest Growth (2023 vs 2022))**

```python
query_5 = """
WITH annual_subcat_profit AS (
    -- Step 1: Base Data (profit sum as per Sub-category & Year)
    SELECT
        product_sub_category,
        EXTRACT(YEAR FROM order_date) AS order_year,
        SUM(net_profit) AS total_profit
    FROM fact_orders
    GROUP BY 1, 2
),
profit_pivot AS (
    -- Step 2: Pivoting (Convert Rows data into column)
    SELECT
        product_sub_category,
        SUM(CASE WHEN order_year = 2022 THEN total_profit ELSE 0 END) AS profit_2022,
        SUM(CASE WHEN order_year = 2023 THEN total_profit ELSE 0 END) AS profit_2023
    FROM annual_subcat_profit
    GROUP BY product_sub_category
)
-- Step 3: Math & Filtering (apply Growth calculate and find  Top 1)
SELECT
    product_sub_category,
    ROUND(profit_2022, 2) AS profit_2022,
    ROUND(profit_2023, 2) AS profit_2023,
    ROUND((profit_2023 - profit_2022), 2) AS absolute_profit_growth
FROM profit_pivot
ORDER BY absolute_profit_growth DESC
LIMIT 3;
"""

# Run query
df_res5 = db_conn.execute(query_5).df()
df_res5
```
